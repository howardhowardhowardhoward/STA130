A simple linear regression is a linear line of best fit used to model the trend of the data:

$$y = \beta_1x + \beta_0 + \epsilon$$

$\beta_1$ - The slope of this line represents the relationship between the variable on the x axis and the variable on the y axis. It depicts the change in the y value when the x value is changed. 

$\beta_0$ - The y-intercept of the linear regression line represents the starting point of the y values. 

$\epsilon$ - The error term shows the difference between the populations data points and the linear regression's predicted data points on the y axis. The error term accounts for randomness in the data. 

The linear regression line is connected to the normal distribution in the sense that if we take any data point from the linear regression line, the error terms to that data point will be normally distributed.

We will now illustrate an example of a linear regression line on a dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Generate synthetic dataset
np.random.seed(0)  # For reproducibility
x = np.random.rand(100) * 10  # 100 random x values from 0 to 10
y = 2.5 * x + np.random.randn(100) * 2  # y = 2.5x + noise

# Step 2: Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Define the linear regression line
def linear_regression_line(x):
    return slope * x + intercept

# Step 3: Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data points', color='blue', alpha=0.5)
plt.plot(x, linear_regression_line(x), color='red', label='Regression line')
plt.title('Linear Regression Demonstration')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to create a synthetic dataset and plot linear regression
def plot_linear_regression(x, y, title):
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

    # Define the linear regression line
    def linear_regression_line(x):
        return slope * x + intercept

    # Create the plot
    plt.scatter(x, y, label='Data points', color='blue', alpha=0.5)
    plt.plot(x, linear_regression_line(x), color='red', label='Regression line')
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid()

# Set up the figure
plt.figure(figsize=(15, 5))

# 1. Positive and large slope
np.random.seed(0)
x1 = np.random.rand(100) * 10
y1 = 5 * x1 + np.random.randn(100) * 2  # Large positive slope
plt.subplot(1, 3, 1)  # 1 row, 3 columns, 1st subplot
plot_linear_regression(x1, y1, 'Large Positive Slope')

# 2. Positive and small slope
x2 = np.random.rand(100) * 10
y2 = 0.5 * x2 + np.random.randn(100) * 2  # Small positive slope
plt.subplot(1, 3, 2)  # 1 row, 3 columns, 2nd subplot
plot_linear_regression(x2, y2, 'Small Positive Slope')

# 3. Negative slope
x3 = np.random.rand(100) * 10
y3 = -2 * x3 + np.random.randn(100) * 2  # Negative slope
plt.subplot(1, 3, 3)  # 1 row, 3 columns, 3rd subplot
plot_linear_regression(x3, y3, 'Negative Slope')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

# Define parameters and generate data
n, x_min, x_range, beta0, beta1, sigma = 20, 5, 5, 2, 3, 5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
Y = beta0 + beta1 * x + errors

# Create DataFrame
df = pd.DataFrame({'x': x, 'y': Y})

# Fit the linear regression model using statsmodels
model_data_specification = smf.ols("y~x", data=df) 
fitted_model = model_data_specification.fit() 

# Add data to legend
df['Data'] = 'Data'

# Plot scatter with trendline
fig = px.scatter(df, x='x',  y='y', color='Data', 
                 trendline='ols', title='Y vs. x')

# Add the fitted trendline explicitly
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

# Plot the theoretical line
x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=f"{beta0} + {beta1} * x", 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['y'][i]],
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)

# Add horizontal line at y-bar
fig.add_scatter(x=x_range, y=[df['y'].mean()]*2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

# Display the plot
fig.show(renderer="png")
```

The OLS method for generating a line of best fit, generates a linear regression line which minimizes the squared difference between the observed values and the predicted values on the line (the residuals).

1. **Linearity**: The relationship between the independent and dependent variables is linear.

3. **Homoscedasticity**: The variance of the residuals is constant across all levels of the independent variable.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm

# Load the Old Faithful dataset
old_faithful = sns.load_dataset('geyser')

# Calculate observed mean difference
observed_mean_diff = old_faithful.groupby('kind')['duration'].mean().diff().iloc[-1]

# Permutation test
def permutation_test(data, n_iterations=10000):
    combined = np.concatenate([data[data['kind'] == 'short']['duration'], 
                                data[data['kind'] == 'long']['duration']])
    
    permuted_diffs = []
    for _ in range(n_iterations):
        np.random.shuffle(combined)
        permuted_short = combined[:len(data[data['kind'] == 'short'])]
        permuted_long = combined[len(data[data['kind'] == 'short']):]
        permuted_diff = np.mean(permuted_long) - np.mean(permuted_short)
        permuted_diffs.append(permuted_diff)
    
    return observed_mean_diff, permuted_diffs

# Perform the permutation test
observed_diff, permuted_differences = permutation_test(old_faithful)

# Calculate p-value
p_value = np.mean(np.abs(permuted_differences) >= np.abs(observed_diff))
print(f"Observed mean difference: {observed_diff}")
print(f"P-value: {p_value}")
```

```python
# Bootstrap confidence interval
def bootstrap_ci(data, n_iterations=10000):
    bootstrapped_diffs = []
    for _ in range(n_iterations):
        bootstrapped_short = data[data['kind'] == 'short'].sample(n=len(data[data['kind'] == 'short']), replace=True)
        bootstrapped_long = data[data['kind'] == 'long'].sample(n=len(data[data['kind'] == 'long']), replace=True)
        
        bootstrapped_diff = bootstrapped_long['duration'].mean() - bootstrapped_short['duration'].mean()
        bootstrapped_diffs.append(bootstrapped_diff)
    
    # Calculate confidence interval
    ci_lower, ci_upper = np.quantile(bootstrapped_diffs, [0.025, 0.975])
    return ci_lower, ci_upper

# Perform the bootstrap confidence interval
ci_lower, ci_upper = bootstrap_ci(old_faithful)
print(f"95% Bootstrap Confidence Interval for the difference in means: ({ci_lower}, {ci_upper})")
```

a) The permutation test is a non-parametric method that evaluates the statistical significance of the observed mean difference between two groups ("short" and "long" wait times) by simulating the null hypothesis of no difference. It begins by calculating the observed mean difference, then combines and shuffles the group labels to create a null distribution of mean differences. The p-value is computed based on the proportion of shuffled differences that are as extreme as the observed difference. In contrast, the bootstrap confidence interval method estimates the sampling distribution of the mean difference by repeatedly resampling the observed data (with replacement) from each group. This results in a distribution of mean differences from which empirical quantiles are calculated to construct a confidence interval, providing a range of plausible values for the true difference between population means.

b) Both methods aim to assess mean differences between groups, yet they differ significantly from the indicator variable model approach. The regression model treats "kind" as a categorical predictor, relying on parametric assumptions about the data, while permutation and bootstrap methods are non-parametric and do not assume normality or homoscedasticity. The permutation test focuses on obtaining a p-value for the observed difference, whereas the bootstrap method emphasizes estimating the uncertainty around the mean difference through confidence intervals. The indicator variable model provides point estimates and statistical inference through established methods, while permutation and bootstrap techniques involve computational simulations to derive results, offering more robustness in cases where assumptions may not hold. Each approach has its strengths, and the choice depends on the analysis context and requirements.
