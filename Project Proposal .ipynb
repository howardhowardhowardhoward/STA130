## Analysis #1

**Research Question:**  
Which form of social connection is most beneficial to one’s mental well-being?

**Explored Variables:**  
In this analysis, I will examine variables related to social connections, specifically the frequency of communication with family, coworkers, friends, and hobby partners, and explore their correlation with wellness variables such as self-rated mental health and self-rated happiness. I plan to visualize these relationships using scatterplots, with the x-axis representing each communication variable and the y-axis representing each wellness variable. This will help illustrate potential correlations.

**Planned Analysis:**  
I will use linear regression to model the relationship between each social variable and each mental health variable. Using Ordinary Least Squares (OLS), I will construct regression lines, and Pearson coefficients will be used to assess the strength of the correlation between each social and mental health variable. By comparing these coefficients, I aim to identify which type of social connection is most closely associated with good mental health and happiness.

**Possible Results of Analysis:**  
I predict that communication with family will show the strongest correlation with good mental health and happiness, while communication with coworkers will show the weakest correlation. This expectation is based on the idea that coworker relationships tend to be more formal and less personal compared to family connections.

**Connection to Course Project Objectives:**  
The project objective is to explore the relationship between social connections and overall well-being. This analysis aligns with this objective by examining the impact of various social connections on mental well-being.

## Analysis #2

**Research Question:**  
To what extent do hugs impact one’s mental well-being?

**Explored Variables:**  
This analysis will focus on the relationship between the frequency of hugs and the level of happiness. I plan to visualize this relationship using side-by-side box plots, with each plot representing a different frequency of hugs and showing the degree of happiness for each frequency level. This visualization will help identify which frequency of hugs is most associated with happiness.

**Planned Analysis:**  
To test whether hugs affect mental well-being, I will conduct hypothesis testing with a significance level of 0.05. I will calculate the median happiness levels for individuals who receive daily hugs versus those who receive no hugs. Using simulation, I will create a sampling distribution of the difference in median happiness levels under the null hypothesis. By calculating the proportion of values as extreme as the observed difference, I will determine the p-value and compare it to the significance level to either accept or reject the null hypothesis.

**Possible Results of Analysis:**  
I anticipate finding a p-value below the significance level, suggesting that the frequency of hugs does impact mental well-being.

**Connection to Course Project Objectives:**  
This analysis supports the project objective of examining the link between social connections and well-being by focusing on hugs as a form of physical and social connection.

## Analysis #3

**Research Question:**  
Which form of virtual communication is associated with the lowest levels of loneliness?

**Explored Variables:**  
This analysis aims to compare various forms of virtual communication, including letters or emails, group video chats, video chats, text messages, and phone calls, to determine which one is associated with the lowest levels of loneliness. The relationship will be visualized through scatterplots, with the x-axis representing the frequency of each communication type and the y-axis showing loneliness levels.

**Planned Analysis:**  
I plan to use linear regression to analyze the correlation between each communication variable and loneliness. By applying OLS and calculating Pearson coefficients, I will determine the strength of each relationship and compare them to identify which communication method is most closely correlated with lower loneliness levels.

**Possible Results of Analysis:**  
I predict that phone calls will yield the highest Pearson coefficient for reducing loneliness, as they are closest to in-person interaction.

**Connection to Course Project Objectives:**  
This analysis contributes to understanding how different types of social connections impact mental well-being by examining the role of virtual communication.
